# Converting to a PyTorch Dataset



```python
import torch
from torch.utils.data import Dataset
```

```python
import os
import pandas as pd
from torchvision.io import read_image

class CustomImageDataset(Dataset):
    def __init__(self, annotations_file, img_dir, transform=None, target_transform=None):
        self.img_labels = pd.read_csv(annotations_file)
        self.img_dir = img_dir
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
        image = read_image(img_path)
        label = self.img_labels.iloc[idx, 1]
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        return image, label
```

```python
class PDE_Dataset(Dataset):
    def __init__(self, hugging_face_dataset):
        self.dataset_dir = hugging_face_dataset['train'].data.to_pandas()['image']
        self.chunk_paths = self.dataset_dir.apply(lambda x: x['path']).tolist()

        self.chunk_info = []
        cumulative = 0
        for path in self.chunk_paths:
            with h5py.File(path, 'r') as f:
                num_samples = f['initial_states'].shape[0]
                self.chunk_info.append((path, cumulative, cumulative + num_samples))
                cumulative += num_samples
        self.total_samples = cumulative

    def __len__(self):
        return self.total_samples

    def __getitem__(self, id):
        for (path, start, end) in self.chunk_info:
            if start <= id < end:
                sample_id_in_chunk = id - start

                with h5py.File(path, 'r') as f:
                    initial_state = f['initial_states'][sample_id_in_chunk]
                    final_state = f['final_states'][sample_id_in_chunk]
                    parameters = f['parameters'][sample_id_in_chunk]
                    elapsed_time = f['elapsed_times'][sample_id_in_chunk]

                return {
                    'initial_state': torch.from_numpy(initial_state).float(),
                    'final_state': torch.from_numpy(final_state).float(),
                    'parameters': torch.from_numpy(parameters).float(),
                    'elapsed_time': torch.tensor(elapsed_time).float()
                }
```

## Loading One Chunk

```python
def create_tf_dataset(h5_data, batch_size=32):
    # Stack all data from chunks
    initial_states = h5_data['initial_states'][:]
    parameters = h5_data['parameters'][:]
    elapsed_times = h5_data['elapsed_times'][:]
    final_states = h5_data['final_states'][:]

    # Create input tuple
    inputs = (initial_states, parameters, elapsed_times)

    # Create TF dataset
    dataset = tf.data.Dataset.from_tensor_slices((inputs, final_states))
    dataset = dataset.shuffle(1000).batch(batch_size)

    return dataset
```

## Loading all chunks

```python
def load_all_chunks(dataset_df):
   all_initial_states = []
   all_parameters = []
   all_elapsed_times = []
   all_final_states = []

   for _, row in dataset_df.iterrows():
       with h5py.File(row['image']['path'], 'r') as f:
           all_initial_states.append(f['initial_states'][:])
           all_parameters.append(f['parameters'][:])
           all_elapsed_times.append(f['elapsed_times'][:])
           all_final_states.append(f['final_states'][:])

   return {
       'initial_states': np.concatenate(all_initial_states),
       'parameters': np.concatenate(all_parameters),
       'elapsed_times': np.concatenate(all_elapsed_times),
       'final_states': np.concatenate(all_final_states)
   }
```

## Loading using Generators

```python
def create_generator_dataset(dataset_df, batch_size=32):
    def generator():
        for _, row in dataset_df.iterrows():
            with h5py.File(row['image']['path'], 'r') as f:
                for i in range(len(f['initial_states'])):
                    yield {
                        'initial_state': f['initial_states'][i],
                        'parameters': f['parameters'][i],
                        'elapsed_time': f['elapsed_times'][i]
                    }, f['final_states'][i]

    return tf.data.Dataset.from_generator(
        generator,
        output_signature=(
            {
                'initial_state': tf.TensorSpec(shape=(200,), dtype=tf.float32),
                'parameters': tf.TensorSpec(shape=(3,), dtype=tf.float32),
                'elapsed_time': tf.TensorSpec(shape=(), dtype=tf.float32)
            },
            tf.TensorSpec(shape=(200,), dtype=tf.float32)
        )
    ).batch(batch_size)
```

```python
df = dataset['train'].data.to_pandas()
tf_dataset = create_generator_dataset(df)
```

# Problem Statement

There are 3 Types of PIN's we could establish.  
- Forward Modeling (Predict the Final State)
- Increment/Operator Learning (Predict at a given time, learn the _time-evolution operator_)
- Parameter Estimation  



# Forward Modeling

The inputs for our PIN will be
- Temperature Profile ($u(x, t_0)$)
- PDE Parameters ($\alpha, k, t_{env}$)
- Time Interval ($\Delta t$)

The output would consist of the final temperature profile
- $\hat{u}(x, t_0 + \Delta t$)

```python
# A helper function to randomly initialize weights and biases
# for a dense neural network layer
def random_layer_params(m, n, key, scale=1e-2):
  w_key, b_key = random.split(key)
  return scale * random.normal(w_key, (n, m)), scale * random.normal(b_key, (n,))

# Initialize all layers for a fully-connected neural network with sizes "sizes"
def init_network_params(sizes, key):
  keys = random.split(key, len(sizes))
  return [random_layer_params(m, n, k) for m, n, k in zip(sizes[:-1], sizes[1:], keys)]

layer_sizes = [784, 512, 512, 10]
step_size = 0.01
num_epochs = 8
batch_size = 128
n_targets = 10
params = init_network_params(layer_sizes, random.key(0))
```
